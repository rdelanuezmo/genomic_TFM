# Miramos de los datos de interacciones negativas de test empleado en el modelo, para cuántas los cuatro modelos predicen interacciones negativas.

```python
test3 = pd.read_csv('/content/drive/MyDrive/test_predictions_with_df_concatenado3.csv')
test4 = pd.read_csv('/content/drive/MyDrive/test_predictions_with_df_concatenado4.csv')
test5 = pd.read_csv('/content/drive/MyDrive/test_predictions_with_df_concatenado5.csv')
test6 = pd.read_csv('/content/drive/MyDrive/test_predictions_with_df_concatenado6.csv')
```

```python
test3.rename(columns={'Predicted': 'pred3'}, inplace=True)
test4.rename(columns={'Predicted': 'pred4'}, inplace=True)
test5.rename(columns={'Predicted': 'pred5'}, inplace=True)
test6.rename(columns={'Predicted': 'pred6'}, inplace=True)
```

```python
test3 = test3[test3['clas'] == -1]
test4 = test4[test4['clas'] == -1]
test5 = test5[test5['clas'] == -1]
test6 = test6[test6['clas'] == -1]
```

```python
df_merged = pd.merge(test3, test4, on=['gene_id', 'interactant_id', 'clas'], how='inner')
df_merged = pd.merge(df_merged, test5, on=['gene_id', 'interactant_id', 'clas'], how='inner')
df_merged = pd.merge(df_merged, test6, on=['gene_id', 'interactant_id', 'clas'], how='inner')
```

```python
condition = (df_merged['pred3'] == -1) & (df_merged['pred4'] == -1) & (df_merged['pred5'] == -1) & (df_merged['pred6'] == -1)
iguales = df_merged[condition]
```
